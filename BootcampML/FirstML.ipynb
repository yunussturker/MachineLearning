# Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
datas = pd.read_csv('datas.txt')
```

```python
datas
```

```python
datas[['boy', 'kilo']]
```

```python
missing_datas = pd.read_csv('missingDatas.txt')
```

```python
missing_datas
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
Yas = datas.iloc[:, 1:4].values
Yas
```

```python
imputer = imputer.fit(Yas[:, 1:4])
```

```python
Yas[:, 1:4] = imputer.transform(Yas[:, 1:4])
Yas
```

```python
ulke = datas.iloc[:,0:1].values
ulke
```

```python
le = preprocessing.LabelEncoder()
```

```python
ulke[:,0] = le.fit_transform(datas.iloc[:,0])
ulke
```

```python
ohe = preprocessing.OneHotEncoder()
```

```python
ulke = ohe.fit_transform(ulke).toarray()
ulke
```

```python
sex = datas.iloc[:,-1:].values
sex
```

```python
sex[:,-1] = le.fit_transform(datas.iloc[:,-1])
sex
```

```python
sex = ohe.fit_transform(sex).toarray()
sex
```

```python
res1 = pd.DataFrame(data=ulke, index = range(22), columns = ['fr', 'tr', 'us'])
res1
```

```python
res2 = pd.DataFrame(data=Yas, index = range(22), columns = ['boy', 'kilo', 'yas'])
res2
```

```python
cinsiyet = datas.iloc[:,-1].values
cinsiyet
```

```python
res3 = pd.DataFrame(data=sex[:,:1], index = range(22), columns = ['cinsiyet'])
res3
```

```python
r1 = pd.concat([res1, res2], axis=1)
r1
```

```python
r2 = pd.concat([r1, res3], axis=1)
r2
```

```python
x_train, x_test, y_train, y_test = train_test_split(r1, res3, test_size=0.33, random_state=0)
x_train, x_test, y_train, y_test
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(x_train)
X_train
```

```python
X_test = sc.fit_transform(x_test)
X_test
```
